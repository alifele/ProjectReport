```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
levelList = []
newList = [[1,1]]
levelList.append(newList.copy())
for i in range(15):
    currentList = newList.copy()
    newList = []
    for segment in currentList:
        alpha = np.random.random()
        lambda_ = np.random.random()/2 + 1/2
        r_1 = 1/(1+alpha**3)**(1/3)*segment[0]
        r_2 = alpha*r_1
        l = segment[0]*lambda_
        newList.append([r_1,l])
        newList.append([r_2,l])
    levelList.append(newList.copy())
```

```python
_=plt.hist(np.array(levelList[10])[:,0])
# _=plt.hist(np.array(levelList[10])[:,1])
plt.grid()
plt.xlabel("R")
plt.ylabel("Segments Count")
```
