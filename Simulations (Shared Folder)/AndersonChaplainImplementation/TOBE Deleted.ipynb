```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phi(x):
    return 6*x - 6*x**2

def iterate(x,n):
    results = []
    for i in range(n):
        results.append(x)
        x = phi(x)

    return np.array(results)
```

```python
x = 1
result = iterate(x,10)
```

```python
plt.plot(result)
```
