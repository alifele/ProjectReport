```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
import numpy as np
def F(X, t, alpha, beta,gamma):
    n, rho = X
    dXdt = [alpha/beta*n +rho - n*rho, gamma/beta**2 * n - rho]
    return dXdt

def vectorField(N,Rho,alpha,beta,gamma):
    U = alpha/beta*N + Rho - N*Rho
    V = gamma/beta**2 * N - Rho
    
    return U,V
```

```python
alpha = 1
beta = 0.5
gamma = 0.4
X0 = [1,1]
t = np.linspace(0, 10, 500)

xList = np.linspace(0,5,100)
N,Rho = np.meshgrid(xList,xList)
```

```python
sol = odeint(F, X0, t, args=(alpha,beta,gamma))
U,V = vectorField(N,Rho,alpha,beta,gamma)
```

```python
p_02 = np.array([(alpha*beta+gamma)/gamma, (alpha*beta+gamma)/beta**2])
```

```python
plt.plot(sol[:,0],sol[:,1],lw=3)
plt.streamplot(N,Rho,U,V)
plt.plot(p_02[0],p_02[1],"o",)
plt.xlabel(r"$n\ [1/A^2]$", fontsize="12")
plt.ylabel(r"$\rho\ [L/A^2]$", fontsize=12)
title = f"Phase Diagram\n" \
        f"α={alpha}, β={beta}, γ={gamma}"

plt.title(title)
plt.grid()
```
